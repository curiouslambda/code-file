```python
from transformers import pipeline
```

Loading the model

```python
model_path = "/mnt/nas2/kikiputri/ner/models/arspraxia_ner_indobert-base_1125_jsonl_filtered_ent-lvl"
classifier = pipeline("ner", model=model_path, aggregation_strategy="simple")
```

Run model with one input

```python
text = "Drama ' High School Love On ' telah merilis beberapa potongan gambar yang menampilkan ChoA Crayon Pop \
    dalam balutan seragam sekolah ."
```

```python
classifier(text)
```

Run model with multiple inputs

```python
texts = [
    "Drama ' High School Love On ' telah merilis beberapa potongan gambar yang menampilkan ChoA Crayon Pop dalam balutan seragam sekolah .",
    "Jin Young B1A4 yang akan berperan sebagai cucu Na Moon Hee . Film ini dijadwalkan untuk rilis pada bulan Januari mendatang ."
]
```

```python
classifier(texts)
```
