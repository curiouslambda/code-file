### 한 폴더안에 모든 json파일이 있을 경우

```python
import os
import json
import pandas as pd
from collections import defaultdict

# 폴더 경로 설정
folder_path = './가공완료_몇개만/article_1121/json_0/'

# 결과를 저장할 딕셔너리 초기화
pub_subj_count = defaultdict(int)
pub_date_count = defaultdict(int)
entities_count = defaultdict(int)
word_count_distribution = defaultdict(int)
entity_classes_count = defaultdict(int)
total_entities = 0
total_word_count = 0

# 폴더 내의 모든 JSON 파일 처리
total_files = 0
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        total_files += 1
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8-sig') as file:
            data = json.load(file)
            pub_subj = data['Pub_Subj']
            pub_date = data['Pub_date']
            sentences = data['data']

            # 결과에 값을 추가
            pub_subj_count[pub_subj] += 1
            # pub_date_count[pub_date] += 1
            pub_date_count[pub_date.split('-')[0]] += 1

            for sentence in sentences:
                entities = sentence['Entities']
                word_count = int(sentence['Word_Count'])
                word_count_distribution[word_count] += 1
                total_word_count += 1
                
                for entity in entities:
                    entity_value = entity['entity']
                    entity_class = entity['entityClass']
                    entities_count[entity_value] += 1
                    entity_classes_count[entity_class] += 1
                    total_entities += 1
                    

# 비율 계산 함수
def calculate_percentage(count_dict, total):
    return {key: f'{value / total * 100:.2f}%' for key, value in count_dict.items()}

# 비율 계산
pub_subj_percentages = calculate_percentage(pub_subj_count, total_files)
pub_date_percentages = calculate_percentage(pub_date_count, total_files)
entities_percentages = calculate_percentage(entities_count, total_entities)
word_count_percentages = calculate_percentage(word_count_distribution, total_word_count)
entity_classes_percentages = calculate_percentage(entity_classes_count, total_entities)

# 데이터프레임 생성
pub_subj_df = pd.DataFrame({'Pub_Subj 비율': pub_subj_percentages, 'Pub_Subj 개수': pub_subj_count})
pub_date_df = pd.DataFrame({'Pub_Date 비율': pub_date_percentages, 'Pub_Date 개수': pub_date_count})
entities_df = pd.DataFrame({'Entity 비율': entities_percentages, 'Entity 개수': entities_count})
word_count_df = pd.DataFrame({'Word_Count 비율': word_count_percentages, 'Word_Count 개수': word_count_distribution})
entity_classes_df = pd.DataFrame({'Entity 클래스 비율': entity_classes_percentages, 'Entity 클래스 개수': entity_classes_count})

# 결과 출력
print('Pub_Subj 데이터프레임:')
print(pub_subj_df)
print(sum(pub_subj_df['Pub_Subj 비율'].str.rstrip('%').astype(float)))

print('\nPub_Date 데이터프레임:')
print(pub_date_df)
print(sum(pub_date_df['Pub_Date 비율'].str.rstrip('%').astype(float)))

print('\nEntities 데이터프레임:')
print(entities_df)
print(sum(entities_df['Entity 비율'].str.rstrip('%').astype(float)))

print('\nWord_Count 데이터프레임:')
print(word_count_df)
print(sum(word_count_df['Word_Count 비율'].str.rstrip('%').astype(float)))

print('\nEntity 클래스 데이터프레임:')
print(entity_classes_df)
print(sum(entity_classes_df['Entity 클래스 비율'].str.rstrip('%').astype(float)))
```

### 압축해제한 폴더가 한 폴더 안에 있을 경우

```python
import os
import json
import pandas as pd
from collections import defaultdict

total_word_count = 0
total_entities = 0

def process_json_files(file_path, pub_subj_count, pub_date_count, entities_count,
                        word_count_distribution, entity_classes_count):

    if file_path.endswith('.json') and not file_path.endswith('statistic.json'):
        with open(file_path, 'r', encoding='utf-8-sig') as file:
            data = json.load(file)
            pub_subj = data['Pub_Subj']
            pub_date = data['Pub_date']
            sentences = data['data']

            # 결과에 값을 추가
            pub_subj_count[pub_subj] += 1
            # pub_date_count[pub_date] += 1
            pub_date_count[pub_date.split('-')[0]] += 1

            for sentence in sentences:
                entities = sentence['Entities']
                word_count = int(sentence['Word_Count'])
                word_count_distribution[word_count] += 1
                global total_word_count
                # global total_entities
                total_word_count += 1
                # total_entities += 1
                for entity in entities:
                    entity_value = entity['entity']
                    entity_class = entity['entityClass']
                    entities_count[entity_value] += 1
                    entity_classes_count[entity_class] += 1

                    #global total_word_count
                    global total_entities
                    #total_word_count += 1
                    total_entities += 1

def calculate_percentage(count_dict, total):
    return {key: f'{value / total * 100:.2f}%' for key, value in count_dict.items()}

```

```python
from tqdm import tqdm

# 폴더 경로 설정
base_folder = './가공완료_몇개만/'

# 결과를 저장할 딕셔너리 초기화
pub_subj_count = defaultdict(int)
pub_date_count = defaultdict(int)
entities_count = defaultdict(int)
word_count_distribution = defaultdict(int)
entity_classes_count = defaultdict(int)
total_entities = 0
total_files = 0
total_word_count = 0

# 폴더 순회 및 JSON 파일 처리
for root, dirs, files in os.walk(base_folder):
    for filename in tqdm(files):
        if filename.endswith('.json'):
            total_files += 1
            file_path = os.path.join(root, filename)
            process_json_files(file_path, pub_subj_count, pub_date_count, entities_count,
                                word_count_distribution, entity_classes_count)

# 비율 계산
pub_subj_percentages = calculate_percentage(pub_subj_count, total_files)
pub_date_percentages = calculate_percentage(pub_date_count, total_files)
entities_percentages = calculate_percentage(entities_count, total_entities)
word_count_percentages = calculate_percentage(word_count_distribution, total_word_count)
entity_classes_percentages = calculate_percentage(entity_classes_count, total_entities)

# 데이터프레임 생성
pub_subj_df = pd.DataFrame({'Pub_Subj 비율': pub_subj_percentages, 'Pub_Subj 개수': pub_subj_count})
pub_date_df = pd.DataFrame({'Pub_Date 비율': pub_date_percentages, 'Pub_Date 개수': pub_date_count})
entities_df = pd.DataFrame({'Entity 비율': entities_percentages, 'Entity 개수': entities_count})
word_count_df = pd.DataFrame({'Word_Count 비율': word_count_percentages, 'Word_Count 개수': word_count_distribution})
entity_classes_df = pd.DataFrame({'Entity 클래스 비율': entity_classes_percentages, 'Entity 클래스 개수': entity_classes_count})

# 결과 출력
print('Pub_Subj 데이터프레임:')
print(pub_subj_df)
print(sum(pub_subj_df['Pub_Subj 비율'].str.rstrip('%').astype(float)))

print('\nPub_Date 데이터프레임:')
print(pub_date_df)
print(sum(pub_date_df['Pub_Date 비율'].str.rstrip('%').astype(float)))

print('\nEntities 데이터프레임:')
print(entities_df)
print(sum(entities_df['Entity 비율'].str.rstrip('%').astype(float)))

print('\nWord_Count 데이터프레임:')
print(word_count_df)
print(sum(word_count_df['Word_Count 비율'].str.rstrip('%').astype(float)))

print('\nEntity 클래스 데이터프레임:')
print(entity_classes_df)
print(sum(entity_classes_df['Entity 클래스 비율'].str.rstrip('%').astype(float)))
```
